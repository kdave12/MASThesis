```python
!pip install be_great
import pandas as pd
from be_great import GReaT
from sklearn.datasets import load_breast_cancer
import os
import torch
torch.cuda.empty_cache()
os.environ['CUDA_LAUNCH_BLOCKING'] = "1"
```

```python
breastCancer = pd.read_csv('breastcancer_real_labeled_without_explicit_indices.csv')
breastCancer
```

```python
model2 = GReaT(llm='distilgpt2', batch_size=16, epochs=100) # pass in fp16=True
model2.fit(breastCancer)
synthetic_data = model2.sample(n_samples=569, max_length=20000)
synthetic_data.to_csv("../beGreatBreastCancer.csv", sep=',', index=False, encoding='utf-8')
```
